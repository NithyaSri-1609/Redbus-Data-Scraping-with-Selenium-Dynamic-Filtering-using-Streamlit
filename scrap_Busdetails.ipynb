#1-APSRTC(Andhra Pradesh State Road Transport Corporation)

```python
# Set up the WebDriver
driver1 = webdriver.Chrome()

# Open the target URL
driver1.get('https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile')
driver1.maximize_window()

# Define the WebDriverWait
wait = WebDriverWait(driver1, 20)

def AP_link(xpath):
    Ap_routelink = []
    Ap_routename = []

    for i in range(1, 6):  # Adjust range as needed
        try:
            # Find all links with the specified XPath
            links = driver1.find_elements(By.XPATH, xpath)
            for link in links:
                Ap_routelink.append(link.get_attribute('href'))
                Ap_routename.append(link.text)

            # Find the next button
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()={i+1}]')
            time.sleep(3)
            
            try:
                next_button.click()
                time.sleep(3)  # Wait for the next page to load
            except ElementClickInterceptedException:
                print(f"Element click intercepted at page {i}, retrying with scroll...")
                driver1.execute_script("window.scrollBy(0, 200);")
                time.sleep(2)  # Wait for the scrolling to finish
                next_button.click()
                time.sleep(3)  # Wait for the next page to load

        except NoSuchElementException:
            print(f"No more pages to paginate at step {i}")
            break

    return Ap_routelink, Ap_routename


Ap_routelink, Ap_routename = AP_link(".//*[@class='route']")

# Create DataFrame and print results
df_A = pd.DataFrame({"Route_Name": Ap_routename, "Route_Link": Ap_routelink})
df_A




```

```python

# Initialize WebDriver
driver = webdriver.Chrome()

# Load route links and names from CSV
df = pd.read_csv(r'F:\NITHYA ONLINE DATA SCIENCE\PROJECT REDBUS\state1_routedetails.csv')

# Lists to store scraped data
data= {
    'Bus Name': [], 'Bus Type': [], 'Departure time': [], 'Arrival time': [], 'Total Duration': [],
    'Rating': [], 'Price': [], 'Seats Available': [], 'Route Name': [], 'Route Link': []
}

# Function to scrape bus details
def scrape_bus_details(link, route_name):
    driver.get(link)
    time.sleep(2)
    
    # Click "View Buses" button if present
    try:
        driver.find_element(By.XPATH, "//div[@class='button']").click()
        time.sleep(2)
    except:
        pass
    
    # Scroll down to load more buses
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    
    # Extract bus details
    try:
        bus_names = driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
        bus_types = driver.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
        Departure_time = driver.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
        Arrival_time = driver.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")
        total_durations = driver.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")
        ratings = driver.find_elements(By.XPATH, "//div[@class='rating-sec lh-24']")
        prices = driver.find_elements(By.XPATH, '//*[@class="fare d-block"]')
        seats = driver.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")
        
        for i in range(len(bus_names)):
            data['Bus Name'].append(bus_names[i].text)
            data['Bus Type'].append(bus_types[i].text if i < len(bus_types) else '')
            data['Departure time'].append(Departure_time[i].text if i < len(Departure_time) else '')
            data['Arrival time'].append(Arrival_time[i].text if i < len(Arrival_time) else '')
            data['Total Duration'].append(total_durations[i].text if i < len(total_durations) else '')
            data['Rating'].append(ratings[i].text if i < len(ratings) else '')
            data['Price'].append(prices[i].text if i < len(prices) else '')
            data['Seats Available'].append(seats[i].text if i < len(seats) else '')
            data['Route Name'].append(route_name)
            data['Route Link'].append(link)
    except Exception as e:
        print(f"Error extracting details for route {route_name}: {e}")

# Scrape details for each route
for _, row in df.iterrows():
    scrape_bus_details(row["Route_Link"], row["Route_Name"])

# Save to CSV
pd.DataFrame(data).to_csv('APSRTC_BUSDETAILS', index=False)

driver.quit()

df_1=pd.read_csv('APSRTC_BUSDETAILS.csv')
df_1


```

```python

# Initialize WebDriver
driver = webdriver.Chrome()

# Load route links and names from CSV
df = pd.read_csv(r'F:\NITHYA ONLINE DATA SCIENCE\PROJECT REDBUS\state9_routedetails.csv')

# Lists to store scraped data
data= {
    'Bus Name': [], 'Bus Type': [], 'Departure time': [], 'Arrival time': [], 'Total Duration': [],
    'Rating': [], 'Price': [], 'Seats Available': [], 'Route Name': [], 'Route Link': []
}

# Function to scrape bus details
def scrape_bus_details(link, route_name):
    driver.get(link)
    time.sleep(2)
    
    # Click "View Buses" button if present
    try:
        driver.find_element(By.XPATH, "//div[@class='button']").click()
        time.sleep(2)
    except:
        pass
    
    # Scroll down to load more buses
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(5)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    
    # Extract bus details
    try:
        bus_names = driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
        bus_types = driver.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
        Departure_time = driver.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
        Arrival_times = driver.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")
        total_durations = driver.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")
        ratings = driver.find_elements(By.XPATH, "//div[@class='rating-sec lh-24']")
        prices = driver.find_elements(By.XPATH, '//*[@class="fare d-block"]')
        seats = driver.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")
        
        for i in range(len(bus_names)):
            data['Bus Name'].append(bus_names[i].text)
            data['Bus Type'].append(bus_types[i].text if i < len(bus_types) else '')
            data['Departure time'].append(Departure_time[i].text if i < len(Departure_time) else '')
            data['Arrival time'].append(Arrival_times[i].text if i < len(Arrival_times) else '')
            data['Total Duration'].append(total_durations[i].text if i < len(total_durations) else '')
            data['Rating'].append(ratings[i].text if i < len(ratings) else '')
            data['Price'].append(prices[i].text if i < len(prices) else '')
            data['Seats Available'].append(seats[i].text if i < len(seats) else '')
            data['Route Name'].append(route_name)
            data['Route Link'].append(link)
    except Exception as e:
        print(f"Error extracting details for route {route_name}: {e}")

# Scrape details for each route
for _, row in df.iterrows():
    scrape_bus_details(row["Route_Link"], row["Route_Name"])

# Save to CSV
pd.DataFrame(data).to_csv('UPSRTC_BUSDETAILS.CSV', index=False)

driver.quit()

df_9=pd.read_csv('UPSRTC_BUSDETAILS.CSV')
df_9

```
